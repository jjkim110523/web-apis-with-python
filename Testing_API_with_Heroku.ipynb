```python
import requests
import json

URL = "https://flask-test-for-study.herokuapp.com/"
response = requests.get(URL)
response.raise_for_status
```

```python
json.loads(response.content.decode( "utf-8" ))
```

```python
file = { "image" : open( "sample.jpg" , "rb" )}
headers = { "type" : "multipart/image" }
filter = "emboss"
response = requests.post(f"{URL}/{filter}", headers=headers, files=file)
response.raise_for_status
```

```python
from PIL import Image
import io
image = Image.open(io.BytesIO(response.content))
image.save( "response_from_heroku.jpg" , "JPEG" )
image
```
